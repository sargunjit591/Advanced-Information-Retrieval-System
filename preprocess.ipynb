```python
import os
import string
import nltk

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

def preprocess_text(text,filename,num):
    if(num==1 or num==100 or num==300 or num==600 or num==800):
        with open('sample_output.txt', 'a', encoding='utf-8') as file:
            file.write(f"\nContents of {filename} before preprocessing:{text}\n")
    # Lowercase 
    text = text.lower()
    if(num==1 or num==100 or num==300 or num==600 or num==800):
        with open('sample_output.txt', 'a', encoding='utf-8') as file:
            file.write(f"\nContents of {filename} after lowering:{text}\n")
    # Tokenization
    tokens = word_tokenize(text)
    if(num==1 or num==100 or num==300 or num==600 or num==800):
        with open('sample_output.txt', 'a', encoding='utf-8') as file:
            file.write(f"\nContents of {filename} after tokenization:{tokens}\n")
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]
    if(num==1 or num==100 or num==300 or num==600 or num==800):
        with open('sample_output.txt', 'a', encoding='utf-8') as file:
            file.write(f"\nContents of {filename} after removing stopwords:{tokens}\n")

    # Remove punctuations
    tokens = [token for token in tokens if token not in string.punctuation]
    if(num==1 or num==100 or num==300 or num==600 or num==800):
        with open('sample_output.txt', 'a', encoding='utf-8') as file:
            file.write(f"\nContents of {filename} after removing punctuations:{tokens}\n")
    # Remove blank space tokens
    tokens = [token for token in tokens if token.isalnum()]

    # Join tokens back into text
    processed_text = ' '.join(tokens)

    if(num==1 or num==100 or num==300 or num==600 or num==800):
        with open('sample_output.txt', 'a', encoding='utf-8') as file:
            file.write(f"\nContents of {filename} after preprocessing:{processed_text}\n\n\n\n")

    return processed_text

def preprocess_file(file_path,filename,num):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
        processed_text = preprocess_text(text,filename,num)


    with open(file_path, 'w', encoding='utf-8') as file:
        file.write(processed_text)

def preprocess_dataset(dataset_path):
    num=1
    for filename in os.listdir(dataset_path):
        if filename.endswith(".txt"):
            file_path = os.path.join(dataset_path, filename)
            preprocess_file(file_path,filename,num)
            num+=1


dataset_path = 'text_files'
with open('sample_output.txt', 'w', encoding='utf-8') as file:
    file.write("Contents of sample text files before and after preprocessing:\n")
preprocess_dataset(dataset_path)
```
